6.2_Advanced Text classification with logistic regression

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
texts=["I love this movie","I hate this film","I hate this film","I love comic movies"]
labels=[1,0,1,1]
vectorizer=TfidfVectorizer()
X = vectorizer.fit_transform(texts)
model=LogisticRegression()
model.fit(X,labels)
text=vectorizer.transform(["I love comic movies"])
prediction=model.predict(text)
print("Predictions:",prediction[0])
```

# 8.Context Free Grammars(CFG)

8.1_Simple CFG parsing

```python
import nltk
grammar = nltk.CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'the'
N -> 'dog' | 'cat'
V -> 'chased' | 'saw'
""")

parser = nltk.ChartParser(grammar)
sentence = ['the', 'dog', 'chased', 'the', 'cat']
for tree in parser.parse(sentence):
    print(tree)
```

8.2_Advanced CFG with Ambiguity resolution

```python
grammar = nltk.CFG.fromstring("""
S -> NP VP
NP -> Det N | N
VP -> V NP
Det -> 'the'
N -> 'dog' | 'cat'
V -> 'chased'
""")

parser = nltk.ChartParser(grammar)
sentence = ['the', 'cat', 'chased', 'dog']
for tree in parser.parse(sentence):
    print(tree)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion.")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
